Criamos uma coluna vazia no dataframe

```python
df['Resultado'] = ''
```

Define a metrica 

```python
df.loc[df['Outcome'] == 1, 'Resultado'] = 'Diabetico'
df.loc[df['Outcome'] == 0, 'Resultado'] = 'No Diabetico'
```

Criamos um grafico de countplot

```python
sns.countplot(data=df, x='Resultado')
```
